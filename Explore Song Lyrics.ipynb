```python
def clean_text_round1(text):
    text = re.sub('\n', '.', text)
    return text
round1 = lambda x: clean_text_round1(x)

songAttributes['Lyrics'] = pd.DataFrame(songAttributes['Lyrics'].apply(round1))
songAttributes  
```

```python
pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity
```

```python
songAttributes['Polarity'] = songAttributes['Lyrics'].apply(pol)
songAttributes['Subjectivity'] = songAttributes['Lyrics'].apply(sub)
```
